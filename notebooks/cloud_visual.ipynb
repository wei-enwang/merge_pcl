```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def plot_cloud(arr_4d):
    # Utility function for plotting a pointcloud
    n, m = arr_4d.shape
    if m == 3:
        arr_4d = np.hstack((arr_4d, np.reshape(['b' for _ in range(n)], (n,1))))
    
    fig = plt.figure()
    fig.suptitle(f"Number of points in this cloud:{n}")
    ax = fig.add_subplot(projection='3d')

    ax.scatter3D(arr_4d[:,0].astype(np.float), 
                 arr_4d[:,1].astype(np.float), 
                 arr_4d[:,2].astype(np.float), 
                 c=arr_4d[:,3], s=0.01)
    plt.show()
```

```python
plt.ion()
```

Load the pointcloud data...

```python
data_path = os.path.join(os.getcwd(), "../src/clouds/")
```

```python
for file in os.listdir(data_path):
    if file.endswith(".npy"):
        cloud = np.load(os.path.join(data_path, file))
        plot_cloud(cloud)
        
```
